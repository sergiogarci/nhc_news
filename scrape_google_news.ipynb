```python
from GoogleNews import GoogleNews
from bs4 import BeautifulSoup
import requests
import csv
```

```python
hazard = ["huracan"] #["deslizamiento","sismo","huracan","lluvia","Falla tectónica","Erupción volcánica","Granizo","Incendio industrial"]
impact = ["Muertos","Heridos","Enfermos","Desaparecidos","Damnificados","Evacuados","Reubicados","Afectados"]
```

```python
#Develop different keywords
keywords = []
for h in hazard:
    for i in impact:
        keyword = "\"" + h + "\"" + " & " + "\"" + i + "\""
        keywords.append(keyword)
print(keywords)
```

```python
#Set up google news preference
googlenews = GoogleNews()
googlenews = GoogleNews(lang='espanol', region='guatemala') #select language and region
googlenews = GoogleNews(period = '3y')
#googlenews = GoogleNews(encode='utf-8')
#Search for each keyword
newsResults = []
date = []
title = []
media = []
description = []
link = []
links = []

for k in hazard: #change hazard to keywords to include impacts 
    googlenews.search(k) #helper = True, scraping_bee = 'BSXBBFSGHFQQAEJ2UQARYTHXP8KLKBIOL30AIVR6Q9VNINFJWZG20OCIQ2GDHYGHKM67Q5361HTF6JPM')
    newsResults = newsResults + googlenews.results(sort = True)
    links = links + googlenews.get_links()
    for c in range(2,50): #Change no of pages
        googlenews.page_at(c)
        newsResults = newsResults + googlenews.results(sort = True)
        links = links + googlenews.get_links()

#Store results
date = [nr['datetime'] for nr in newsResults]
title = [nr['title'] for nr in newsResults]
media = [nr['media'] for nr in newsResults]
description = [nr['desc'] for nr in newsResults]
link = [nr['link'] for nr in newsResults]

#Clear List
googlenews.clear()
```

```python
#Download html from individual pages
headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/70.0.3538.77 Safari/537.36"}
session = requests.Session()
html_text = []
for l in links:
    if "youtube" in l:
        html_text.append("")
        continue
    elif "carscoops" in l:
        html_text.append("")
        continue
    elif "wsoctv" in l:
        html_text.append("")
        continue
    elif "auto" in l:
        html_text.append("")
        continue
    elif "car" in l:
        html_text.append("")
        continue
    elif "motor" in l:
        html_text.append("")
        continue
    elif "driv" in l:
        html_text.append("")
        continue
    else:
        s = session.get(l, headers = headers)
        html = BeautifulSoup(s.text, 'html.parser')
        html_text.append(html.get_text())
        print(l)
```

```python
#Store csv
header = ['date','title','media','description','link','html'] #Headers
#Prepare data for writing 
data = []
for i in range(0,len(newsResults)):
    dt = [date[i],title[i],media[i],description[i],link[i],html_text[i]]
    data.append(dt)

#write to csv
with open(r'D:\GA Bensi\newsresults_huracan.csv', 'w', encoding='UTF8', newline='') as f:
    writer = csv.writer(f)
    writer.writerow(header)
    writer.writerows(data)
```
